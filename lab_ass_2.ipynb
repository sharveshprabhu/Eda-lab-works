```python
import pandas as pd
import numpy as np
```

```python
df_a = pd.DataFrame({
    'StudentID': [1, 3, 5, 7, 9],
    'Score': [89, 39, 50, 97, 22]
})

df_b = pd.DataFrame({
    'StudentID': [2, 4, 6, 8, 10],
    'Score': [98, 93, 44, 77, 69]
})

v_align = pd.concat([df_a, df_b], ignore_index=True)
print("Vertical Concatenation:")
print(v_align)

h_align = pd.concat([df_a, df_b], axis=1)
print("\nHorizontal Concatenation:")
print(h_align)
```

```python
data_se = pd.DataFrame({
    'StudentID': [2, 4, 6, 8, 10, 12],
    'ScoreSE': [98, 93, 44, 77, 69, 56]
})

data_ml = pd.DataFrame({
    'StudentID': [1, 2, 3, 4, 5, 6, 8, 10],
    'ScoreML': [39, 93, 49, 44, 55, 78, 97, 87]
})

match_inner = pd.merge(data_se, data_ml, on='StudentID', how='inner')
print("\nInner Join:")
print(match_inner)

match_left = pd.merge(data_se, data_ml, on='StudentID', how='left')
print("\nLeft Join:")
print(match_left)

match_right = pd.merge(data_se, data_ml, on='StudentID', how='right')
print("\nRight Join:")
print(match_right)

match_outer = pd.merge(data_se, data_ml, on='StudentID', how='outer')
print("\nOuter Join:")
print(match_outer)
```

```python
orders = pd.DataFrame({
    'Quantity': [100, 250, 300, 150],
    'UnitPrice': [50, 60, 45, 80]
})

orders['TotalPrice'] = orders['Quantity'] * orders['UnitPrice']
print("\nSales Data:")
print(orders)
```

```python
print("\nDescriptive Statistics:")
print(orders.describe())
```

```python
firm_df = pd.DataFrame({
    'Company': ['ABC', 'XYZ', 'ABC', 'PQR', 'XYZ', 'ABC']
})

print("\nCompany Counts:")
print(firm_df['Company'].value_counts())
```

```python
grid = np.arange(15).reshape((3, 5))
weather_df = pd.DataFrame(
    grid,
    index=['Rainfall', 'Humidity', 'Wind'],
    columns=['Bergen', 'Oslo', 'Trondheim', 'Stavanger', 'Kristiansand']
)
print("\nOriginal DataFrame:")
print(weather_df)

stack_res = weather_df.stack()
print("\nStacked Data:")
print(stack_res)

unstack_res = stack_res.unstack()
print("\nUnstacked Data:")
print(unstack_res)
```

```python
seq_1 = pd.Series([0, 111, 222, 333], index=['zeros', 'ones', 'twos', 'threes'])
seq_2 = pd.Series([444, 555, 666], index=['fours', 'fives', 'sixs'])

combined_series = pd.concat([seq_1, seq_2], keys=['Number1', 'Number2'])
print("\nHierarchical Series:")
print(combined_series)

print("\nUnstacked Series:")
print(combined_series.unstack())
```
